```python
def cosine_similarity(merged):
    try:
        merged = str(merged)
        merged = merged.split('!SPLIT!')
        corpus = [merged[0],merged[1]]
        vectorizer = TfidfVectorizer()
        X = vectorizer.fit_transform(corpus)

        v1 = np.array([x  for y in X[0].toarray() for x in y])
        v2 = np.array([x  for y in X[1].toarray() for x in y])

        return np.dot(v1, v2) / (np.sqrt(np.sum(v1**2)) * np.sqrt(np.sum(v2**2)))
    except:
        return 0
```

```python
data_set['cosine'] = data_set['merged'].apply(cosine_similarity)
```

```python
data_set['cosineClean'] = data_set['cleanMerged'].apply(cosine_similarity)
```
